```python
#bai2:
class Date:
    def __init__(self,day,month,year):
        self.d = day
        self.m = month
        self.y = year
    def namnhuan(self):
        return self.y % 400 == 0 or ( self.y % 4 == 0 and self.y % 100 != 0)
    def thang(self):
        if self.m in [1,3,5,7,8,10,12] :
            return 31
        if self.m in [4,6,9,11]:
            return 30
        if self.m == 2:
            return 29 if self.namnhuan() else 28
    def loi(self):
        if self.m <= 0 or self.m > 12:
            return False
        if self.d > self.thang() or self.d <= 0:
            return False
        if self.y <= 0:
            return False
        return True
    def nextday(self):
        self.d += 1
        if self.d > self.thang():
            self.d = 1 
            self.m += 1
            if self.m > 12:
                self.m = 1
                self.y += 1
        return f"{self.d} : {self.m} : {self.y}"
day,month,year = map(int,input().split())
manh = Date(day,month,year)
if not manh.loi():
    print("Ngay thang nam khong hop le")
else:
    print(manh.nextday())
```

```python
#bai6:
class Rectangle():
    def __init__(self,w,h):
        self.w = w
        self.h = h
    def scale(self,k):
        self.w *= k
        self.h *= k
    def perimeter(self):
        return (self.w + self.h) * 2
    def area(self):
        return self.w * self.h
w,h = map(float,input().split())
thanh = Rectangle(w,h)
k = int(input("phóng to k lần : "))
thanh.scale(k)
print(f"{thanh.area():.2f} {thanh.perimeter():.2f}")
```

```python
class BankAccount:
  def __init__(self,money):
    self. m = money
    
  def deposit(self,amount):
    self. m += amount
    return BankAccount(self. m)
    
  def withdraw(self,amount):
    if amount <= self. m:
      self. m -= amount
      return BankAccount(self. m)
    else:
      return BankAccount(self. m)

  def __str__(self):
    return f'{self. m :.2f}'

ha, hat = map(str,input().split())
ht = BankAccount(float(hat))
n = int(input())
i=0
while i < n:
  a, b = map(str,input().split())
  i += 1
  amount = float(b)
  if a == 'DEPOSIT':
    ht.deposit(amount)
  elif a == 'WITHDRAW':
    ht.withdraw(amount)

print(ht)
```

```python
#bài 11:
import math
class Angle:
    def __init__(self, A, B, C, D):
        self.A = A
        self.B = B
        self.C = C
        self.D = D
    def vector(self, P, Q):
        return [Q[0]-P[0], Q[1]-P[1], Q[2]-P[2]]
    def cross(self, a, b):
        return [
            a[1]*b[2] - a[2]*b[1],
            a[2]*b[0] - a[0]*b[2],
            a[0]*b[1] - a[1]*b[0]
        ]
    def dot(self, a, b):
        return a[0]*b[0] + a[1]*b[1] + a[2]*b[2]
    def length(self, v):
        return math.sqrt(self.dot(v, v))
    def angle_between_planes(self):
        AB = self.vector(self.A, self.B)
        AC = self.vector(self.A, self.C)
        BC = self.vector(self.B, self.C)
        BD = self.vector(self.B, self.D)
        n1 = self.cross(AB, AC)
        n2 = self.cross(BC, BD)
        cos_phi = abs(self.dot(n1, n2)) / (self.length(n1) * self.length(n2))
        phi = math.degrees(math.acos(cos_phi))
        return phi
A = list(map(float, input().split()))
B = list(map(float, input().split()))
C = list(map(float, input().split()))
D = list(map(float, input().split()))
goc = Angle(A, B, C, D)
print(f"{goc.angle_between_planes():.2f}")
```

```python
import math

class Complexnumber:
  def __init__(self, real, image):
    self. a = real
    self. b = image
    
  def __add__(self,other):
    new_a = (self. a + other. a)
    new_b = (self. b + other. b)
    return Complexnumber(new_a, new_b)
    
  def __mul__(self, other):
    new_a = self. a * other. a -self. b * other. b
    new_b = self. a * other. b + self. b * other. a
    return Complexnumber(new_a, new_b)
    
  def __abs__(self):
    return math.sqrt((self. a **2 +self. b**2))
    
  def conjugate(self):
    return Complexnumber(self. a, -self. b)
    
  def __str__(self):
    if self. b >= 0:
      return f'{self. a :.2f} + {self. b :.2f}i'
    else:
      return f'{self. a :.2f} - {abs(self. b) :.2f}i'
      
cn1 = input(). split()
z1 = Complexnumber(float(cn1[0]), float(cn1[1]))
cn2 = input().split()
z2 = Complexnumber(float(cn2[0]), float(cn2[1]))
tong = z1 + z2 
tich = z1 * z2
modul = abs(tich)
lien_hop = tich. conjugate()
print(tong)
print(tich)
print(f'{modul :.2f}')
print(lien_hop)  
```

```python
#bai 13:
import math
class Triangle():
    def __init__(self,canh1,canh2,canh3):
        self.a = canh1
        self.b = canh2
        self.c = canh3
    def doublegetArea(self):
        a, b, c = self.a, self.b, self.c
        if a <= 0 or b <= 0 or c <= 0 or a + b <= c or a + c <= b or b + c <= a:
            return -1
        p = (a + b + c) / 2
        return math.sqrt(p * (p - a) * (p - b) * (p - c))
canh1,canh2,canh3 = map(int,input("Nhập ll 3 cạnh của tam giác: ").split())
manh = Triangle(canh1,canh2,canh3)
if manh.doublegetArea() == -1:
    print("invalid")
else:
    print(f"{manh.doublegetArea():.2f}")
        
```
