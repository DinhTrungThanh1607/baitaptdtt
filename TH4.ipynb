```python
#W4A3 
n=int(input())
giai_thua = 1
if n < 0:
    print('Bạn phải nhập số nguyên vào thì mới tính được giai thừa ')
else:
    if n == 0 :
        giai_thua = 1
        print(giai_thua)
    else:
        for i in range (1,n+1):
            giai_thua *= i
        print(giai_thua)
```

```python
#W4A13
a = int(input())
b = int(input())
tong_uoc_a = []
tong_uoc_b = []
for i in range(1,a ):
    if a % i == 0 :
        tong_uoc_a.append(i)
for j in range(1,b):
    if b % j == 0 :
        tong_uoc_b.append(j)
if sum(tong_uoc_a)==b and sum(tong_uoc_b)==a:
    print("True")
else:
    print("False")


```
