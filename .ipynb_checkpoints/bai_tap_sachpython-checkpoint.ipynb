```python
#Dung append() dể thêm các phần tử mới vào list cũ đã tạo , nhưng dùng append thì chỉ được thêm các phần tử đó  vào cuối thôi nên có phần hơi bất tiện so với insert()
khang_vu=['ha vy','diem nhu','hoang duyen','linh chi']
khang_vu.append('thuc uyen')
print(khang_vu)
```

```python
#Dùng insert() để thêm các phần tử mới vào list cũ đã được lập,có thể thêm vào bất kỳ vị trí nào mà mình muốn không nhất thiết phải bắt buộc thêm vào cuối list như là lệnh append()
khang_vu=['ha vy','diem nhu','hoang duyen','linh chi'] # các phần tử trong danh sách đều có vị tri và số thứ tự được đếm theo quy luât tăng dần từ 0,1,2,..,v.v

khang_vu.insert(0,'thuc uyen')
print(khang_vu)
```

```python
# Sử dụng hàm del để xóa đi 1 phần tử nào đó trong 1 list nhừng khi xóa đi phần tử đó thì mình sẽ không thể tìm lại được phần tử đó nữa nên sử dụng lệnh del có phần nào hạn chế hơn về mặt công năng
khang_vu=['ha vy','diem nhu','hoang duyen','linh chi']
print(khang_vu)
del khang_vu[0]
print(khang_vu)
```

```python
# Việc sử dụng lệnh pop() có phần đa dạng về công năng hơn lệnh del ở chỗ ,khi bạn đã xóa 1phaanf tử khỏi cái list đó thì bạn vẫn có thẻ sử dụng lại được phần tử đó được chứ kh như del lệnh del thì sẽ không thể nào mà sử dụng lại được cái phần tử đó mà lênh del thì chỉ có tác dụng để xóa phần tử mình muốn khỏi list 
# Ví dụ khi bạn lâp trình 1 trang web trò chơi khi người đăng kí thoát ra thì bạn phải xóa ngưởi đó khỏi danh sách người đăng ký, khi bạn muốn tìm lại cái dánh sách mình đã xóa thì nó sẽ vẫn còn lưu lại ở trên đoạn code chứ kh bị mất đi
khang_vu=['ha vy','diem nhu','hoang duyen','linh chi']
print(khang_vu)
havyxinhgai=khang_vu.pop(0)
print(khang_vu)
print(f"{havyxinhgai.upper()} ơi anh yêu em nhiều lắm ,dear mr Trới")
```
