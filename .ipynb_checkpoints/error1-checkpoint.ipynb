```python
#Bai 2 
import math
n=0
for _ in range(10000):
    n=int(input("Nhap vao 1 so nguyen duong"))
    if n>0:
        break #Dung để ngắt vòng lặp      continue nếu đạt được rồi thì bảo qua hết
    else:
        print("Ban phai nhap so nguyen duong de dung")
print("So nguyen duong ban nhap duoc la",n)
flag=True
for i in range(2,int(math.sqrt(n))+1):
    if n%i==0:
        flag=False
        break
if flag==True:
    print(n,"la so nguyen to")
else:
    print(n,"khong phai la so nguyen to")
```

```python
try:
    n=int(input())
    if n<0:
        print("ban phai nhap so nguyen dương thì mới tính được giai thừa")
    elif n==1 or n==0:
        print("1")
    else:
        giaithua=1
        for i in range(1,n+1):
            giaithua *=i
    print(giaithua)
except ValueError:
    print("bạn đã bị lỗi")
```

```python
try:
    n = int(input())
    n = abs(n)
    count = 0
    if n==0:
        count = 1
        print(count)
        
    count = 0
    for _ in range(10000):
        if n>0:
            count += 1
            n //= 10
        else:
            break
    print(count)
except ValueError:
    print("ban nhap sai du lieu roi")
```
