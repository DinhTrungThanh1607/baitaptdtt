```python
#16
a=list(map(int,input().split()))
ha={}
for i in a:
    if i in ha:
        ha[i] += 1
    else:
        ha[i] = 1
a=set(a)
res=0
for j in a:
    if ha[j] > j:
        res+=ha[j]-j
    elif ha[j]<j:
        res+= ha[j]
if res==0:
    print("YES")
else:
    print(f'NO {res}')
    
```

```python
#20
def non(n):
    n=str(n)
    return len(n)
ha=list(map(int,input().split()))
n=int(input())
res=0
for i in ha:
    if non(i) == n:
        res += i
print(res)
```

```python
#22
n=int(input())
i=0
ha=[]
while i<n:
    i += 1
    m=input().split()
    ha.append(m[-1])

ha.sort()
hat={}
for j in ha:
    if j in hat:
        hat[j] += 1
    else:
        hat[j] = 1
res=0
ha=set(ha)
ha=list(ha)
for h in ha:
    m=hat[h]
    res=max(m,res)
for key,value in hat.items():
    if hat[key]==res:
        print(key)
        break
```

```python
#25
n=input()
k=int(input())
ha=len(n)-k
hat=[]
for i in n:
   
    while ha>0 and len(hat)>0 and hat[-1]<i:
        hat.pop()
        ha -= 1
    hat.append(i)
print("".join(hat[:k]))
```

```python
#31
hat=list(input())
ha=len(hat)
for i in range(ha):
    ham=hat[i]
    ad=-1
    for j in range(i+1,ha):
        if hat[j]<=ham:
            if hat[j]<ha[i]:
                ham=hat[j]
                ad=j
            elif hat[j]==ham and ad!=-1:
                ad=j
    if ad!=-1:
        if i==0 and hat[ad]=='0':
            continue
        hat[i],hat[ad]=hat[ad],hat[i]
        break
print(''.join(hat))
```
