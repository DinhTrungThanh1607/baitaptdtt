```python
#1
def binary_search(arr,left,right,target):
    if left>right:
        return -1
    mid=(left+right)//2
    if a[mid]==target :
        return mid
    elif a[mid]<target:
        return binary_search(arr,mid+1,right,target)
    else:
        return binary_search(arr,left,mid-1,target)
a=list(map(int,input().split()))
n=int(input())
ha=binary_search(a,0,len(a)-1,n)
print(ha)
```
