```python
#W5A1
def max_of_two(a: int, b: int) -> int:
    return max(a,b)
try:
    a,b = map(int, input().split())
    print(max_of_two(a,b))
except ValueError:
    print("Nhap sai gia tri")
```

```python
#W5A3
import math
x = int(input())
def kiemtrasnt(x):
    if x < 2:
        return False
    for i in range(2, int(math.sqrt(x)) + 1):
        if x % i == 0:
            return False
    return True
if kiemtrasnt(x) == True:
    print(x, "la snt")
else:
    print(x,"khong la snt")
```

```python
#W5A4
n = int(input())
def kiemtrasohoanhao(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s = s + i
    if s == n:
        return True
    return False
if kiemtrasohoanhao(n):
    print(n,"la so hoan hao")
else:
    print(n,"khong la so hoan hao")
```

```python
#W5A5
x,y = map(int, input().split())
def binary(n):
    arr = []
    while n > 0:
        arr.append(n % 2)
        n = n // 2
    arr2 = arr[::-1]
    return arr2
sum = 0
a = binary(x)
b = binary(y)
while len(a) < len(b):
    a.insert(0,0)
while len(b) < len(a):
    b.insert(0,0)
for i in range(len(a)):
    if a[i] != b[i]:
        sum +=1
print(sum)
```

```python
#W5A5               Dung toan tu bitwise XOR: so sanh tung bit neu giong se tra ve 1 khac tra ve 0
x, y = map(int, input().split())

# x ^ y sẽ tìm ra các bit khác nhau
# bin().count('1') sẽ đếm số lượng các bit khác nhau đó
ket_qua = bin(x ^ y).count('1')

print(ket_qua)
```

```python
#W5A10
x,y = (map(str, input().split()))
def kiemtra(x,y):
    if len(x) != len(y):
        return False
    a_to_b = {}
    b_to_a = {}
    for phantu_a, phantu_b in zip(x, y):
        if phantu_a in a_to_b:
            if a_to_b[phantu_a] != phantu_b:
                return False
        else:
            a_to_b[phantu_a] = phantu_b
        if phantu_b in b_to_a:
            if b_to_a[phantu_b] != phantu_a:
                return False
        else:
                b_to_a[phantu_b] = phantu_a
                return False
    return True
if kiemtra(x,y):
    print(True)
else:
    print(False)
    
```
