```python
#4
n=input()
ha=n.split()
dict={}
for i in ha:
    key,value = i.split(':')
    if key in dict:
        dict[key].append(value)
    else:
        dict[key]=[value]
print(dict)
```

```python
#9
import ast
dict1 = ast.literal_eval(input()) 
dict2 = ast.literal_eval(input())
ha=dict1.copy()
for key,value in dict2.items():
    if key in ha :
        ha[key] += value 
    else:
        ha[key] = value
hat=dict(sorted(ha.items()))
print(hat)
```

```python
#13
a=list(map(str,input().split()))
ha=[]
at=[]
for i in range(len(a)):
    if i%2==0:
        ha.append(a[i])
    else:
        at.append(a[i])
hat={}
for x,y in zip(at,ha):
    hat[x]=y
print(hat)
```
