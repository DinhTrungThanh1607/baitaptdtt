```python
#Bai 4
h=int(input())
for tamgiac in range(h):
    print(' '*(h-tamgiac),end='')
    if tamgiac==0:
        print('*')
    elif tamgiac==h-1:
        print('*'*(2*tamgiac+1))
    else:
        print('*'+' '*(2*tamgiac-1)+'*')
```
